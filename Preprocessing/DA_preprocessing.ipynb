# **identify outliers**

```python
from scipy import stats
import numpy as np
outliers_col = []
for col in df.columns:
    if df[col].dtype in ['int64', 'float64']:
        z_scores = np.abs(stats.zscore(df[col]))  # use col variable, not 'col'
        outliers = df[z_scores > 3]
        outliers_col.append(col)              # Z > 3 is considered an outlier
        print(f"Column '{col}' has {outliers.shape[0]} outliers")
print(outliers_col)
print(len(outliers_col))
```

```python
import pandas as pd
import numpy as np
from scipy import stats
from scipy.stats.mstats import winsorize

# --- Step 1: Identify numeric columns with outliers using Z-score ---
outlier_cols = []

for col in df.columns:
    if df[col].dtype in ['int64', 'float64']:
        z_scores = np.abs(stats.zscore(df[col].dropna()))
        if (z_scores > 3).any():   # if any value has |Z| > 3
            outlier_cols.append(col)

print("Columns with outliers:", outlier_cols)

# --- Step 2: Apply Winsorization only on those columns ---
for col in outlier_cols:
    df[col] = winsorize(df[col], limits=[0.10, 0.10])  # caps lowest & highest 10%
    print(f"Winsorization applied on '{col}'")

# --- Optional: Verify Winsorization ---
for col in outlier_cols:
    z_scores = np.abs(stats.zscore(df[col].dropna()))
    print(f"After Winsorization, '{col}' has {(z_scores > 3).sum()} outliers left")
```

# **Scaling and transformation**

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, PowerTransformer

# Columns to transform
cols_to_scale = ['LM_WT']
cols_to_transform = ['Production (MT)', 'ENERGY (Energy Consumption)']

# 1️⃣ Standardize LM_WT
scaler = StandardScaler()
df['LM_WT_std'] = scaler.fit_transform(df[['LM_WT']])

# 2️⃣ Transform Production and ENERGY using Yeo-Johnson (handles zeros & negatives)
pt = PowerTransformer(method='yeo-johnson')

df['Production_MT_pt'] = pt.fit_transform(df[['Production (MT)']])
df['ENERGY_pt'] = pt.fit_transform(df[['ENERGY (Energy Consumption)']])

# Check new columns
print(df[['LM_WT', 'LM_WT_std', 'Production (MT)', 'Production_MT_pt', 'ENERGY (Energy Consumption)', 'ENERGY_pt']].head())
```
