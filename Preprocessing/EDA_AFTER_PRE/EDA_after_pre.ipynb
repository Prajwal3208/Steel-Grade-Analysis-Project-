```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('pre_DA_DATA_final.xlsx')
```

```python
def generate_eda_summary(df):
    summary = []

    for col in df.columns:
        col_data = df[col]
        dtype = col_data.dtype

        if pd.api.types.is_numeric_dtype(col_data):
            # Handle numeric columns
            min_val = col_data.min() if col_data.notnull().any() else np.nan
            max_val = col_data.max() if col_data.notnull().any() else np.nan
            mean_val = col_data.mean() if col_data.notnull().any() else np.nan
            median_val = col_data.median() if col_data.notnull().any() else np.nan
            mode_val = col_data.mode().iloc[0] if not col_data.mode().empty else np.nan
            std_val = col_data.std() if col_data.notnull().any() else np.nan
            var_val = col_data.var() if col_data.notnull().any() else np.nan
            skew_val = col_data.skew() if col_data.notnull().any() else np.nan
            kurt_val = col_data.kurtosis() if col_data.notnull().any() else np.nan
            range_val = max_val - min_val if col_data.notnull().any() else np.nan

        else:
            # Handle categorical or non-numeric columns
            min_val = "-"
            max_val = "-"
            mean_val = "-"
            median_val = "-"
            std_val = "-"
            var_val = "-"
            skew_val = "-"
            kurt_val = "-"
            range_val = "-"
            mode_val = col_data.mode().iloc[0] if not col_data.mode().empty else np.nan

        summary.append({
            'Column Name': col,
            'Data Type': dtype,
            'Min': min_val,
            'Max': max_val,
            'Mean': mean_val,
            'Median': median_val,
            'Mode': mode_val,
            'Standard Deviation': std_val,
            'Variance': var_val,
            'Range': range_val,
            'Skewness': skew_val,
            'Kurtosis': kurt_val
        })

    return pd.DataFrame(summary)


EDA = generate_eda_summary(df)


EDA.to_excel("EDA_after_pre.xlsx", index=False)
print("EDA summary saved as 'EDA_after_pre.xlsx'")
```
